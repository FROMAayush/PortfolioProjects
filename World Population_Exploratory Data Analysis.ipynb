```python
#Lets look correlation between numeric values, say population
df.corr()
#Its giving error because AFG has non numeric data type
```

```python
#So, lets filter the non-numeric column
df_numeric = df.select_dtypes(include = ['number']) #selects only numberic columns
#Lets compute correlation of only numbers now
df_corr = df_numeric.corr()
df_corr
```

```python
#Lets visulaize using heat map
sns.heatmap(df_numeric.corr(), annot = True)
plt.show()
```

```python
#Lets amke the graph more clear and numbers more visible
sns.heatmap(df_numeric.corr(), annot = True)
plt.rcParams['figure.figsize'] = (15,12)
plt.show()
```

```python
#Lets see if there are any continents that have grown faster in population than others
df_numeric.groupby(df['Continent']).mean().sort_values (by="2022 Population", ascending = False)
```

```python
#Let's visualize this data
df2 = df_numeric.groupby(df['Continent']).mean().sort_values (by="2022 Population", ascending = False)
df2.plot()
```

```python
#For ease in visualization, Lets transpose Population data sets to row and continent sets to columns
df3 = df2.transpose()
```

```python
df3.plot()
```

```python
#The graph above still doesnt seem easy to visualize. There are information that we donot need
#like density, area etc. which is not required for us to visulize the continent with highest population growth
#Let's visualize this data
df4 = df_numeric.groupby('Continent')[
    ['2022 Population', '2020 Population', '2015 Population', 
     '2010 Population', '2000 Population', '1990 Population', 
     '1980 Population', '1970 Population']
].mean().sort_values(by="2022 Population", ascending=False)
```

```python
#The error KeyError: 'Continent' is suggesting that the column 'Continent' does not exist in df_numeric.
#Next 3-4 steps are used for fixing the errors, setting 'Continent' back as a column
df_numeric = df_numeric.reset_index()

print(df_numeric.columns)
```

```python
# Let's see if 'Continent' exists in the original df
print(df.head()) 
```

```python
#If 'Continent' is in df but missing from df_numeric, 
#we probably dropped it when selecting numeric columns.
#Fix it by manually adding it back:
df_numeric['Continent'] = df['Continent']
```

```python
# Now trying the Groupby Code Again
df4 = df_numeric.groupby('Continent')[
    ['2022 Population', '2020 Population', '2015 Population', 
     '2010 Population', '2000 Population', '1990 Population', 
     '1980 Population', '1970 Population']
].mean().sort_values(by="2022 Population", ascending=False)
df4
```

```python
#using more shortcut approach for the same:
df.columns
df4 = df_numeric.groupby('Continent')[df.columns[5:13]].mean().sort_values(by="2022 Population", ascending = False)
df4
```

```python
df5 = df4.transpose()
df5
```

```python
#This seemed little opposite in descending order, lets do in ascending order. 
df44 = df_numeric.groupby('Continent')[
    ['1970 Population', '1980 Population', '1990 Population', 
     '2000 Population', '2010 Population', '2015 Population', 
     '2020 Population', '2022 Population']
].mean().sort_values(by="2022 Population", ascending=False)
df44
```
